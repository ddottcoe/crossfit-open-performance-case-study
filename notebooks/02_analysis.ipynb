# 02 - Analysis

This notebook loads the derived tables and reproduces the main portfolio charts.

```
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

root = Path('..')
athlete = pd.read_csv(root / 'data' / 'derived' / 'athlete_year_open_poc.csv')
fit = pd.read_csv(root / 'data' / 'derived' / 'open_event_training_fit_summary.csv')
corr = pd.read_csv(root / 'data' / 'derived' / 'open_event_training_fit_correlations.csv')
```

```
years = athlete[['year', 'overall_rank']].dropna().sort_values('year')
plt.figure(figsize=(8,4.5))
plt.plot(years['year'], years['overall_rank'], marker='o')
plt.gca().invert_yaxis()
plt.title('CrossFit Open Overall Rank by Year')
plt.xlabel('Year')
plt.ylabel('Overall Rank (lower is better)')
plt.grid(True, alpha=0.3)
plt.show()
```

```
fit[['year', 'workout_code', 'performance_vs_year_rank_pct', 'event_fit_score']].sort_values('performance_vs_year_rank_pct').head(10)
```

- The stronger story is that transferability matters more than raw movement exposure.

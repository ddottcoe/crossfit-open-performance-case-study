# 03 - Training Plan

This notebook translates the analysis into an actionable Open-prep training structure.

## Weakness buckets

- Wall walk / inversion under fatigue
- Snatch cycling in conditioning
- Thruster and clean receiving under fatigue
- Burpee-to-implement transitions
- Squat-based movement under breathing disruption

## Seasonal structure

**April to August**
- aerobic base
- pulling volume
- lower-body strength
- low-fatigue inversion exposure

**September to November**
- weakness-biased development
- snatch cycling
- wall walk density
- burpee + implement combinations

**December to January**
- Open-specific conversion
- short-to-medium race pieces
- benchmark retests

**February to March**
- sharpening and pacing
- 1 to 2 Open-style tests per week
